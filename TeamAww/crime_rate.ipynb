```python
# 5년 전후 대비 발생량 기울기 파악 

rape_a = rape_happend['2019']/rape_happend['2015']
rob_a = rob_happend['2019']/rob_happend['2015']
violence_a = violence_happend['2019']/violence_happend['2015']
heist_a = heist_happend['2019']/heist_happend['2015']
murder_a = murder_happend['2019']/murder_happend['2015']

crime_a = pd.DataFrame({
    '강간': rape_a, 
    '살인': murder_a,
    '절도': rob_a,
    '강도': heist_a,
    '폭력': violence_a , 
    }) ; crime_a 
```

```python
for j in range(5):
    cnt =0
    for i in range(25): 
        if crime_a.values[i][j] >= 1:
            cnt += 1
    print( '2015년 대비 2019년의 "', crime_a.columns[j],'" 범죄가 증가한 구의 수:',cnt, ' |  감소한 구의 수:', 25-cnt , ' => 범죄가 증가한 구의 비율:' , round( cnt / 25 , 3)  )
```
