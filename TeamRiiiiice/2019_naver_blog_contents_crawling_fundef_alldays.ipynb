## 02. 데이터 수집 및 저장

```python
for dal in range(9,13):  # 9월부터 12월까지

    # 01. 페이지 준비(9월~ 12월 설정)
    if dal == 9:    
        url = 'https://search.naver.com/search.naver?where=blog&query=%EC%8C%80%20-%EC%97%85%EC%B2%B4%2C%20-%EC%A0%9C%EA%B3%B5%EB%B0%9B%2C%20-%EC%A7%80%EC%9B%90%EB%B0%9B%2C%20-%EC%9B%90%EA%B3%A0%EB%A3%8C%2C%20-%EB%AC%B4%EC%83%81%2C%20-%ED%98%91%EC%B0%AC%2C%20-%EC%86%8C%EC%A0%95%EC%9D%98%2C%20-%EC%82%AC%EB%A3%8C%2C%20-%EA%B3%A0%EB%9E%98%EB%B0%A5%2C%20-%EA%B8%B8%EB%83%A5%EC%9D%B4%2C%20-%EA%B3%A0%EC%96%91%EC%9D%B4%2C%20-%EA%B0%95%EC%95%84%EC%A7%80&sm=tab_opt&nso=so%3Ar%2Cp%3Afrom20190901to20190930'
    if dal == 10:   
      url = 'https://search.naver.com/search.naver?where=blog&query=%EC%8C%80%20-%EC%97%85%EC%B2%B4%2C%20-%EC%A0%9C%EA%B3%B5%EB%B0%9B%2C%20-%EC%A7%80%EC%9B%90%EB%B0%9B%2C%20-%EC%9B%90%EA%B3%A0%EB%A3%8C%2C%20-%EB%AC%B4%EC%83%81%2C%20-%ED%98%91%EC%B0%AC%2C%20-%EC%86%8C%EC%A0%95%EC%9D%98%2C%20-%EC%82%AC%EB%A3%8C%2C%20-%EA%B3%A0%EB%9E%98%EB%B0%A5%2C%20-%EA%B8%B8%EB%83%A5%EC%9D%B4%2C%20-%EA%B3%A0%EC%96%91%EC%9D%B4%2C%20-%EA%B0%95%EC%95%84%EC%A7%80&sm=tab_opt&nso=so%3Ar%2Cp%3Afrom20191001to20191031'
    if dal == 11:    
      url = 'https://search.naver.com/search.naver?where=blog&query=%EC%8C%80%20-%EC%97%85%EC%B2%B4%2C%20-%EC%A0%9C%EA%B3%B5%EB%B0%9B%2C%20-%EC%A7%80%EC%9B%90%EB%B0%9B%2C%20-%EC%9B%90%EA%B3%A0%EB%A3%8C%2C%20-%EB%AC%B4%EC%83%81%2C%20-%ED%98%91%EC%B0%AC%2C%20-%EC%86%8C%EC%A0%95%EC%9D%98%2C%20-%EC%82%AC%EB%A3%8C%2C%20-%EA%B3%A0%EB%9E%98%EB%B0%A5%2C%20-%EA%B8%B8%EB%83%A5%EC%9D%B4%2C%20-%EA%B3%A0%EC%96%91%EC%9D%B4%2C%20-%EA%B0%95%EC%95%84%EC%A7%80&sm=tab_opt&nso=so%3Ar%2Cp%3Afrom20191101to20191130'
    if dal == 12:    
      url = 'https://search.naver.com/search.naver?where=blog&query=%EC%8C%80%20-%EC%97%85%EC%B2%B4%2C%20-%EC%A0%9C%EA%B3%B5%EB%B0%9B%2C%20-%EC%A7%80%EC%9B%90%EB%B0%9B%2C%20-%EC%9B%90%EA%B3%A0%EB%A3%8C%2C%20-%EB%AC%B4%EC%83%81%2C%20-%ED%98%91%EC%B0%AC%2C%20-%EC%86%8C%EC%A0%95%EC%9D%98%2C%20-%EC%82%AC%EB%A3%8C%2C%20-%EA%B3%A0%EB%9E%98%EB%B0%A5%2C%20-%EA%B8%B8%EB%83%A5%EC%9D%B4%2C%20-%EA%B3%A0%EC%96%91%EC%9D%B4%2C%20-%EA%B0%95%EC%95%84%EC%A7%80&sm=tab_opt&nso=so%3Ar%2Cp%3Afrom20191201to20191231'

    browser = webdriver.Chrome("./chromedriver")
    browser.get(url)
    time.sleep(1)


    #스크롤을 끝까지 내려주는 코드
    while True:
        for w in range(1015):
            browser.find_element_by_css_selector("html").send_keys(Keys.END)
        break


    # 최종 수집리스트
    #제목 / 날짜
    titles = []
    dates = []
    #내용
    contents = []
    #id
    ids = []
    #댓글
    chats = []
    chats2 = []


    title = browser.find_elements_by_css_selector("a.api_txt_lines.total_tit")
    date = browser.find_elements_by_css_selector("span.sub_time.sub_txt")

    cnt = 0
    while cnt<= 1050 :
        try:

            titles.append(title[cnt].text)
            dates.append(date[cnt].text)

            title[cnt].click()

            time.sleep(4)

            browser.switch_to.window(browser.window_handles[-1])

            try:

                browser.switch_to.frame('mainFrame')
                #content
                try:
                    blogcontent = browser.find_elements_by_css_selector(
                        "div.se-main-container")
                    for i in blogcontent:
                        contents.append(i.text)
                    if blogcontent == []:
                        blogcontent = browser.find_elements_by_css_selector(
                            "div#postViewArea")
                        for i in blogcontent:
                            contents.append(i.text)
                    if blogcontent == []:
                        contents.append(np.nan)  
                except:
                    contents.append(np.nan)

                
                #id
                try:        
                    ids.append(browser.find_element_by_css_selector("span.itemfont.col").text)
                except:
                    ids.append(np.nan)


                #chat

                try:
                    browser.find_element_by_css_selector(
                        "div.area_comment.pcol2").click()
                    time.sleep(1.5)
                    blogchat = browser.find_elements_by_css_selector(
                        "span.u_cbox_contents")
                    midchat=[]
                    q = ""

                    for j in blogchat:
                        q += j.text +  " || "
                        midchat.append(j.text)
                    chats.append(q)
                    chats2.append(midchat)
                except:
                    chats.append(np.nan)
                    chats2.append(np.nan)
                
                browser.close()
                browser.switch_to.window(browser.window_handles[0])
            
            except:
                
                browser.close()
                browser.switch_to.window(browser.window_handles[0])

                chats.append(np.nan)
                contents.append(np.nan)
                ids.append(np.nan)
        
        except:
            browser.switch_to.window(browser.window_handles[-1])

        cnt += 1


    #저장
    a=pd.DataFrame(titles,columns=["title"])
    b = pd.DataFrame(dates, columns=["date"])
    c = pd.DataFrame(chats, columns=["chat"])
    d = pd.DataFrame(contents, columns=["content"])
    e = pd.DataFrame(ids,columns=["id"])

    df = pd.concat([a, b, c, d, e], axis=1)
    
    #05.엑셀파일로 저장
    df.to_excel(f'./alldaycrawlblog/naver_blog_conents_ssal_2019_{dal}월.xlsx')

    ##06. 드라이버 종료
    browser.close()
```
