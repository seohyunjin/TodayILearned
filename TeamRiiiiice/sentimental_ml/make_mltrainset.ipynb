# 학습데이터 만들기

## 01. 쿠팡 데이터 모아 합치기

```python
import pandas as pd
```

```python
data1 = pd.read_excel('./train_data/coupang_naver_total/part1.xlsx',header= None)
data2= pd.read_excel('./train_data/coupang_naver_total/part2.xlsx')
data3 = pd.read_excel('./train_data/coupang_naver_total/part3_1.xlsx')
data4 = pd.read_excel('./train_data/coupang_naver_total/part3_2.xlsx')
data5 = pd.read_excel('./train_data/coupang_naver_total/part4.xlsx')
data6 = pd.read_excel('./train_data/coupang_naver_total/part5.xlsx')
```

```python
# train_x 합치기 
all_train_x_df= pd.concat( [data1[0], data2['문장'],data3['문장'],data4['문장'],data5[0],data6[0] ], axis = 0)
all_train_x_df=all_train_x_df.reset_index()
all_train_x_df
```

```python
# train_y합치기
all_train_y_df= pd.concat( [data1[1], data2['감성'],data3['감성'],data4['감성'],data5['0.1'],data6['0.1'] ], axis = 0)
all_train_y_df=all_train_y_df.reset_index()
all_train_y_df
```

```python
# 점수 매기기
num_train_y_df =[]
for i in range(len(all_train_y_df)):
    if all_train_y_df[0][i] == 'neutral':
        num_train_y_df.append(0)
    elif all_train_y_df[0][i] == 'positive':
        num_train_y_df.append(1)
    else:
        num_train_y_df.append(-1)

len( num_train_y_df )
```

```python
coupang_train = pd.DataFrame({ 'train':all_train_x_df[0],
'api': all_train_y_df[0],
'score': num_train_y_df})
coupang_train
```

```python
coupang_train.to_excel('./train_data/coupang_naver_total/total_senti_coupang.xlsx',encoding='utf-8')
```
